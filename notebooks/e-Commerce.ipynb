# Tamanho global

* Market size: USD 16 tri em 2021
    1. B2B domina o mercado, com um share de 63% do montante global
    2. B2C está crescendo, no entanto, com o maior acesso à smartphones e propagandas direcionadas (Google Ads, Facebook Ads)
    3. No B2C, os marketplaces são a modalidade preferida à lojas especializadas
    4. Asia-Pacífico é a maior região do segmento, com 55% do mercado
    
    
* Principais players: Amazon, JD.com, Apple, Alibaba (AliExpress), Flipkart, Walmart, eBay, Best Buy, The Home Depot, Shopee, Wish.com


* Mobile shopping e social commerce (comprar direto de plataformas sociais) são a tendência.


* Na esfera de pagamentos, a trend buy now - pay later com plataformas cross-border está dominando o mercado.
    1. Nelo (fintech de ex-executivos da Uber) levantou investimentos para expandir essa modalidade de serviço na LatAm. 
    
    
* Mais de 2/3 dos consumidores já compraram em sites fora de seu país de origem

# Tamanho Brasil

* Market size: USD 25,3 bi em 2021
    1. O Brasil teve um crescimento de 26,8% no segmento de 2020 pra 2021
    2. O ticket médio foi de BRL 443
    3. Mais de 85mi de consumidores
    


* Principais players: Magazine Luiza, Mercado Livre, Via Varejo, Americanas, Dafiti, Amazon, Privalia, Shopee, AliExpress, Netshoes, Shein, Kabum

# Perspectiva no Brasil

* O crescimento do segmento depende muito do acesso à tecnologia em áreas rurais e mais pobres. (Mais de 70% dos acessos são pelo celular)


* A pandemia deu uma paulada na supply chain mundial. Então esta está em processo de reorganização e modernização. No curtíssimo prazo, isso pode segurar um pouco o crescimento do mercado (até 2023, segundo reports), mas logo resultará em um potencial de crescimento ainda maior.


* A adoção do 5G e a aceitação do metaverso no Brasil poderão ditar o rate de crescimento do segmento.
