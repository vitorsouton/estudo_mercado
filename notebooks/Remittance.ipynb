# Tamanho global

* Market size: USD 17,9bi em 2021. Expectativa de USD 60,5bi em 2030
    1. Contando somente MTOs e Fintechs
    2. Contabilizando-se bancos, USD 702bi em 2020. Expectativa de USD 1.2tri em 2030 





* Principais players: Exchange Solutions, Western Union, MoneyGram, WorldRemit, Azimo, Digital Wallet, PayPal Group (XOOM), TransferGo, Wise, UAE Exchange
    1. O mercado está moderadamente fragmentado, com grandes empresas já consolidadas mas ainda com espaço para inovações e novos players

    



* O mercado vem crescendo devido à adoção de pagamentos mobile-based em moeda digital e à imigração
    1. Vários governos pelo mundo já estão preparando regulamentação específica direcionada ao envio de dinheiro por imigrantes de baixa-renda para seus países de origem
    2. Do outro lado, para países emergentes, o dinheiro advindo dessas remessas são parte importante da economia. Logo a regulação ou não existe ou é bem permissiva.


* Ainda há um temor por parte de países desenvolvidos de que os canais digitais de remittance possam ser utilizados para financiamento de grupos terroristas, lavagem de dinheiro e na operação de drogas ilícitas


* As grandes instituições financeiras estão correndo atrás do prejuízo e diminuindo suas taxas e burocracia, porém os operadores digitais independentes ainda conseguem oferecer taxas mais competitivas
    1. Existe uma pressão do G20 para que as operadoras baixem suas taxas para próximo de 5%

# Tamanho Brasil

* Market size: Movimentação de USD 4,9bi em 2020 (incluindo bancos)
    1. USD 1,5bi de outflow
        * Portugal e EUA como principais destinos (37%)
    2. USD 3,3bi de inflow
        * EUA e Reino Unido como principais fontes (66%)
        

* Principais players: Western Union, MoneyGram, PayPal Group (XOOM), Wise, Travelex Bank, BrasilRemittance
    1. Atualmente há mais de 180 instituições autorizadas pelo BCB para processar transações de FX. Pórem o mercado ainda é dominado pelos bancos tradicionais e grandes MTOs.


* Em 2018, o BCB mudou a regulamentação para receber moeda estrangeira no país, autorizando que instituições brasileiras operem no mercado internacional, convertam a moeda e entreguem ao cliente em BRL
    1. Assim, as conversões são negociadas internacionalmente e o cliente final não negocia câmbio no Brasil
    
    
* A partir de setembro de 2022, IPs poderão atuar em FX


# Perspectiva Brasil

* O cidadão brasileiro é pouco globalizado. Proporcionalmente, o Brasil não ocupa nem o top 20 em movimentações internacionais de dinheiro. 


* Por outro lado, nos últimos anos houve um escape de brasileiros de média e alta renda para a Europa, Canadá e EUA. 


* Com a pressão do G20 para a redução de taxas de remittance, grandes MTOs estão custodiando o dinheiro e faturando mais no câmbio


* Com a liberação do BCB para que IPs atuem na área, é possível a consolidação de parcerias entre MTOs estabelecidas e IPs com caminhos de transferência estabelecidos


* Em valores, a tendência é que este segmento do mercado caia até o Q2 do ano que vem
