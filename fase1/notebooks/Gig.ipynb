# Tamanho global

* O mercado global foi estimado em USD 297bi em 2020. (58% disso representado por aplicativos de mobilidade como Uber e Lyft e cerca de 4% por plataformas de freelance)
    1. É esperado que o mercado global de freelancers cresça constantemente até a próxima década.
    2. A tendência nos países desenvolvidos é que mais da metade da força de trabalho seja de gig-workers até 2030

# Tamanho Brasil

* Estimar o tamanho do mercado no Brasil baseado em faturamento é bem difícil, pois não há dados o suficiente.
* Por número de usuários: 
    1. A plataforma brasileira de freelance, Closeer, viu um crescimento de 60% na base de usuários no primeiro semestre de 2021.
    2. A Workana registrou um aumento explosivo no cadastro de freelancers brasileiros durante a pandemia.
    3. 47% dos trabalhadores no Brasil atuam na modalidade gig (contando freelance, transporte, entrega, etc)
    

* Players no Brasil: 
    1. Plataformas de freelance: Upwork, Workana, Closeer, Toptal, Simply Hired, PeoplePerHour, Aquent, Crowded, Fiverr, 99Designs, Freelancer, Guru
    1. Outras categorias: iFood, 99Taxi, Uber, Rappi, GetNinjas, Airbnb
    
    


# Perspectiva no Brasil

* O mercado de gig-economy está em franco crescimento e tende a continuar assim.
* A grande maioria dos usuários e dinheiro deste segmento se concentram em transporte e entrega, porém as plataformas de freelance estão ganhando mais espaço.
* As grandes plataformas internacionais também atuam no Brasil (Workana, Toptal, Fiverr, etc), conectando empregadores estrangeiros com profissionais brasileiros.
* Em 2021 o Brasil aprovou a criação do visto para nômades digitais, logo é esperado que também ocorra em menor escala o movimento contrário.
