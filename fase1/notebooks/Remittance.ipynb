# Tamanho Brasil

* Market size: Movimentao de USD 4,9bi em 2020 (incluindo bancos)
    1. USD 1,5bi de outflow
        * Portugal e EUA como principais destinos (37%)
    2. USD 3,3bi de inflow
        * EUA e Reino Unido como principais fontes (66%)
        

* Principais players: Western Union, MoneyGram, PayPal Group (XOOM), Wise, Travelex Bank, BrasilRemittance
    1. Atualmente h mais de 180 instituies autorizadas pelo BCB para processar transaes de FX. Prem o mercado ainda  dominado pelos bancos tradicionais e grandes MTOs.


* Em 2018, o BCB mudou a regulamentao para receber moeda estrangeira no pas, autorizando que instituies brasileiras operem no mercado internacional, convertam a moeda e entreguem ao cliente em BRL
    1. Assim, as converses so negociadas internacionalmente e o cliente final no negocia cmbio no Brasil
    
    
* A partir de setembro de 2022, IPs podero atuar em FX
