# Tamanho Brasil

* Market size: USD 1,3bi em 2020. Expectativa de USD 2,2bi em 2023

* Principais players: bet365, Sportingbet, Betano, Betfair, pixbet, Estrelabet, Betnacional, KTO, Blaze Casino



* O Brasil ainda não tem regulamentação específica para este segmento, portanto é vedada a abertura de casas de aposta online no país. Mas isso não impede que plataformas internacionais atuem na área; basta estarem de acordo com a legislação e regulamentação do país aonde esteja sua sede.

* Apostas e jogos de azar são proibidos. As exeções são: Poker, corridas de cavalo em hipódromos, loterias federais e apostas com quota fixa (onde se encaixam as apostas esportivas desde que as odds sejam fixas e pré-estabelecidas no momento da aposta) 

# Perspectiva Brasil

* O mundo todo está de olho no mercado brasileiro, esperando somente a regulamentação definitiva. O país é chamado do "Gigante adormecido das apostas".

* Espera-se que o Brasil possa gerar mais Gross gaming revenue do que todos os países do mundo, exceto EUA e Macau. 

* Apesar de já ser um mercado grande no Brasil, seu futuro é incerto. A legislação que regulamentará a modalidade no país já está tramitando no congresso, porém não se sabe se será permissiva ou proibitiva. O mais provável, pórem, é que seja permitida a abertura de empresas brasileiras no setor.

* O mercado é de especial risco para os processadores de pagamento, pois há um entendimento que estes são responsáveis por eventuais problemas com o dinheiro. Na maioria dos sites de apostas, há uma cláusula nos termos e condições afirmando que eventuais disputas judiciais devem ser resolvidas no país no qual o site está sediado. Mas na prática, alguns juízes entendem que quem faz a movimentação do dinheiro deve ser responsabilizado.
