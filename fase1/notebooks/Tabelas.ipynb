```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(17, 9)})
```

```python
df = pd.read_csv('../data/dados_preliminares.csv', )
```

```python
df_ecommerce = df.iloc[4:7]
df.drop([4, 5, 6], inplace=True)
```

# Comparação tamanho global

```python
global_ = sns.barplot(x=df['segmento'], y=df['tamanho_global_bilhoes'])
global_.set(ylim=(0, 1000));
```

# Comparação estimativa global 2030

```python
global_2030 = sns.barplot(x=df['segmento'], y=df['tamanho_global_2030_est'])
global_2030.set(ylim=(0, 1000));
```

# Comparação tamanho Brasil

```python
brasil = sns.barplot(x=df['segmento'], y=df['tamanho_brasil_bilhoes'])
brasil.set(ylim=(0, 20));
```
