```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size': 21})
```

```python
df = pd.read_csv('../data/profile.csv')
df.head()
```

# Sexo

```python
male = round((df.percentage_male.sum() / df.shape[0]), 3)
female = 1 - male
```

```python
plt.figure(figsize=(17,9))
plt.bar(['male', 'female'], [male, female], color=['red','blue'])
plt.tight_layout();
```

# Idade

```python
idades = df[['age_18-24', 'age_25-34', 'age_35-44', 'age_45plus']]
idade_pcts = round((idades.sum() / idades.shape[0]), 3)
```

```python
plt.figure(figsize=(17,9))
plt.bar(['18-24', '25-34', '35-44', '45+'], idade_pcts, color=['red','blue','green','grey'])
plt.tight_layout();
```

# Duração da visita

```python
plt.figure(figsize=(17,9))
plt.rcParams.update({'font.size': 17})
plt.bar(df.company_name, df.visit_duration_min)
plt.tight_layout();
```
